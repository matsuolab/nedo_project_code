```python
import requests, time
from bs4 import BeautifulSoup
import re

# ベースURLとページ範囲
base_url = "https://eiga.com"
pages = range(1, 11)  # 1から10までのページ

# 空のリストでurlsを初期化
urls = []

# 正規表現パターン
pattern = re.compile(r'/movie/\d+/$')

# 各ページをループ
for page in pages:
    # ページのURLを構築
    url = f"{base_url}/movie/review/ranking/{page}/"

    # URLにアクセスしてHTMLを取得
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # `/movie/数字/` にマッチするリンクを抽出
    for a in soup.find_all('a', href=True):
        if pattern.search(a['href']):
            full_url = base_url + a['href']
            if full_url not in urls:
                urls.append(full_url)  # リンクをリストに追加
    time.sleep(1)

print(urls)

```

```python
import requests, time
from bs4 import BeautifulSoup
import re

# urls リスト（例として手動でいくつかのURLを定義）
# urls = ['https://eiga.com/movie/100801/', 'https://eiga.com/movie/100802/', 'https://eiga.com/movie/100803/']

# 空のリストでurls2を初期化（このステップは今回は不要となるため削除可能）
urls2 = []

# 正規表現パターン
pattern = re.compile(r'/movie/\d+/review/\d+/$')

# ベースURL
base_url = "https://eiga.com"

# 保存先のファイルパス
file_path = '/content/drive/MyDrive/Scraping/input/movie_review_url.txt'

# 各映画ページをループ
for url in urls:
    # 追加のページURLを含める
    review_pages = [url + "review/", url + "review/all/2/", url + "review/all/3/"]

    # 各レビューページをループ
    for review_url in review_pages:
        # レビューページにアクセスしてHTMLを取得
        # print(review_url)
        response = requests.get(review_url)
        soup = BeautifulSoup(response.content, 'html.parser')

        # `/movie/数字/review/数字/` にマッチするリンクを抽出
        for a in soup.find_all('a', href=True):
            if pattern.search(a['href']):
                full_url = base_url + a['href']
                if full_url not in urls2:
                    urls2.append(full_url)  # リンクをリストに追加
                    # ファイルに書き込み
                    with open(file_path, 'a') as file:
                        file.write(full_url + '\n')
        time.sleep(1)  # 各ページアクセス後に少し待つ

print(urls2)
print(len(urls2))
urls2 = list(set(urls2))  # 重複を除去する（このステップは必要に応じて実施）
print(len(urls2))
```

```python
import requests
import json, os, time
from bs4 import BeautifulSoup

# 基本ディレクトリパス（適切に設定してください）
# ROOT_PATH = "/path/to/your/directory/"
output_dir = ROOT_PATH + "jsonl/"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# urls2リスト（事前に定義が必要です、ここでは仮に空としています）
# urls2 = ['https://example.com/movie1', 'https://example.com/movie2']  # 仮のURLを追加

def fetch_all_reviews_to_jsonl(jsonl_filename):
    filename = os.path.join(output_dir, jsonl_filename)
    # 抽出したページ数のカウント
    count = 0
    # 全ページ数のカウント
    total_count = 0

    with open(filename, 'w', encoding='utf_8_sig') as file:
        for url in urls2:
            total_count += 1
            page_url = url
            response = requests.get(page_url)
            soup = BeautifulSoup(response.content, 'html.parser')
            title = soup.title.text if soup.title else ''

            # <span class="rating-star">から評価値を抽出（2つ目の要素）
            rating_spans = soup.find_all('span', class_='rating-star')
            rating = rating_spans[1].get_text(strip=True) if len(rating_spans) > 1 else '評価なし'

            # <div class="txt-block">の内容を抽出
            txt_block = soup.find('div', class_='txt-block')
            text_content = txt_block.get_text(strip=True) if txt_block else ''

            # 最終的なテキスト
            text = f"{rating}点。 {text_content}"

            # JSONL形式で書き込み
            page_data = {'title': title, 'url': page_url, 'text': text}
            json.dump(page_data, file, ensure_ascii=False)
            file.write('\n')

            count += 1
            if count % 100 == 0:
                print(f"Processed {count} pages.")
            time.sleep(1)

        print(f"Total pages retrieved: {count}/{total_count}")

# ファイル名を指定して関数を呼び出し
fetch_all_reviews_to_jsonl('movie_reviews.jsonl')
```
