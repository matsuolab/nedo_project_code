```python
import requests
from bs4 import BeautifulSoup

# URLの指定
url = 'https://sist8.com/ronri'

# URLからHTMLコンテンツを取得
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# すべてのaタグを見つけて、href属性を出力
links = [a.get('href') for a in soup.find_all('a') if a.get('href')]

# リンクを出力
for link in links:
    print(link)
```

```python
import requests
from bs4 import BeautifulSoup, NavigableString, Tag
import json

def extract_section_text(html_content, section_header):
    soup = BeautifulSoup(html_content, 'html.parser')
    results = []
    headers = soup.find_all('h2')

    for header in headers:
        if header.text.strip() == section_header:
            content = []
            current = header.next_sibling
            while current and current.name != 'h2':
                if isinstance(current, (NavigableString, Tag)):
                    text = current.text.strip() if isinstance(current, Tag) else current.strip()
                    if text:
                        content.append(text)
                current = current.next_sibling
            if content:
                results.append(' '.join(content))
    return results

def save_to_jsonl(data, file_path):
    with open(file_path, 'a', encoding='utf-8') as file:
        for entry in data:
            json.dump(entry, file, ensure_ascii=False)
            file.write('\n')

def extract_and_save(url):
    response = requests.get(url)
    if response.status_code == 200:
        problem_texts = extract_section_text(response.content, "問題")
        seikai_texts = extract_section_text(response.content, "正解")
        kaisetsu_texts = extract_section_text(response.content, "解説")
        if problem_texts and seikai_texts:
            soup = BeautifulSoup(response.content, 'html.parser')
            text = f"問題###\n{problem_texts}\n\n正解###\n{seikai_texts}\n\n解説###\n{kaisetsu_texts}"
            entry = {"title": soup.title.text if soup.title else "", "url": url, "text": text}
            save_to_jsonl([entry], '/content/drive/MyDrive/Scraping/jsonl/quiz_2.jsonl')
        else:
            print(f"Some sections are missing in {url}.")
    else:
        print(f"Failed to retrieve {url}. Status code: {response.status_code}")

# URLリストファイルのパス
url_file_path = '/content/drive/MyDrive/Scraping/input/quiz_2_url.txt'

# URLリストを読み込む
urls = []
with open(url_file_path, 'r') as file:
    urls = [line.strip() for line in file if line.strip()]

urls = list(set(urls))

for url in urls:
    extract_and_save(url)
```
