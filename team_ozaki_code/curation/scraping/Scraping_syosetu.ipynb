```python
import shutil
import os

# 元のURLリストファイルのパス
url_file_path = '/content/drive/MyDrive/Scraping/input/syosetu_url_original.txt'

# コピー元のディレクトリパス
source_dir = '/content/drive/MyDrive/Scraping/jsonl/syosetu/'

# コピー先のディレクトリパス
destination_dir = '/content/drive/MyDrive/Scraping/jsonl/syosetu2/'

# コピー先のディレクトリがなければ作成
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# URLファイルを開き、最初の100件のURLを処理
with open(url_file_path, 'r') as file:
    for i, line in enumerate(file):
        if i >= 100:  # 100件まで処理
            break

        # URLからファイル名の部分を抽出
        novel_code = line.strip().split('/')[-2]

        # コピー元とコピー先のファイルパスを構築
        source_file = os.path.join(source_dir, f"{novel_code}.jsonl")
        destination_file = os.path.join(destination_dir, f"{novel_code}.jsonl")

        # ファイルをコピー
        shutil.copy2(source_file, destination_file)

print("Completed copying the specified files.")
```

```python
import os

# 指定されたディレクトリ
directory_path = '/content/drive/MyDrive/Scraping/jsonl/syosetu2'

# 削除する文字列
remove_text = 'しおりを解除しました。\\nブックマークを解除しました。\\nエラーが発生しました。\\nエラーの原因がわからない場合はヘルプセンターをご確認ください。\\n'

# ディレクトリ内の全ファイルを処理
for filename in os.listdir(directory_path):
    # ファイルの完全なパスを構築
    file_path = os.path.join(directory_path, filename)

    try:
        # ファイルを開いて内容を読み込む
        with open(file_path, 'r+', encoding='utf-8') as file:
            content = file.read()

            # 指定した文字列を削除
            new_content = content.replace(remove_text, '')

            # 変更があった場合のみファイルを上書き
            if new_content != content:
                file.seek(0)
                file.write(new_content)
                file.truncate()
                print(f"Updated file: {filename}")
            else:
                print(f"No changes made to: {filename}")

    except Exception as e:
        print(f"Failed to process file {filename}. Error: {e}")
```

```python
import requests
import json
import os
import time
from bs4 import BeautifulSoup
from trafilatura import extract

# 出力ディレクトリの設定
ROOT_PATH = '/content/drive/MyDrive/Scraping/'
input_url_path = ROOT_PATH + "input/syosetu_url.txt"
output_dir = ROOT_PATH + "jsonl/syosetu/"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

headers = {
    'Host': 'ncode.syosetu.com',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
}

def update_url_file(url_file, urls):
    with open(url_file, 'w', encoding='utf-8') as f:
        for url in urls:
            f.write(url + '\n')

def fetch_all_pages_to_jsonl(url_file):
    with open(url_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    processed_urls = []
    for line in lines:
        line = line.strip()
        if ',1' in line:
            processed_urls.append(line)
            continue  # Skip already processed URLs

        base_url, *_ = line.split(',')
        jsonl_filename = f"{base_url.split('/')[-2]}.jsonl"
        filename = os.path.join(output_dir, jsonl_filename)

        with open(filename, 'w', encoding='utf_8_sig') as jsonl_file:
            processed_count = 0  # Initialize the count of processed URLs for current base URL
            for page_number in range(1, 1000):  # Assume a max of 999 pages
                page_url = f"{base_url}{page_number}/"
                response = requests.get(page_url, headers=headers)

                if response.status_code == 200:
                    soup = BeautifulSoup(response.text, 'html.parser')
                    title = soup.find('title').text if soup.find('title') else "No Title"
                    extracted_text = extract(response.content, include_tables=False, target_lang='ja')

                    if not extracted_text or "。" not in extracted_text:
                        continue

                    page_data = {'title': title, 'url': page_url, 'text': extracted_text}
                    json.dump(page_data, jsonl_file, ensure_ascii=False)
                    jsonl_file.write('\n')

                    processed_count += 1
                    if processed_count % 10 == 0:
                        print(f"Processed URL: {page_url} - Title: {title}")

                elif response.status_code == 404:
                    print(f"Page {page_number}: Status 404 Not Found - Stopping")
                    break
                time.sleep(1)  # Sleep to reduce load on server

        processed_urls.append(f"{base_url},1")  # Mark base URL as processed
        update_url_file(url_file, processed_urls)

# 関数を実行
fetch_all_pages_to_jsonl(input_url_path)

```

```python
import requests
from bs4 import BeautifulSoup
import time

headers = {
    'Host': 'yomou.syosetu.com',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
}

def check_pages(base_url, start, end):
    for i in range(start, end + 1):
        url = f"{base_url}{i}/"
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            title = soup.find('title').text if soup.find('title') else "No Title"
            print(f"Page {i}: Status 200 OK, Title: {title}")
        elif response.status_code == 404:
            print(f"Page {i}: Status 404 Not Found")
            break  # 404が出たらループを終了
        else:
            print(f"Page {i}: Status {response.status_code}")
        time.sleep(2)  # サーバーへの負荷軽減のための遅延

# 基本URL
base_url = 'https://ncode.syosetu.com/n0593it/'

# スタートページとエンドページ
start_page = 1
end_page = 500  # 実際の最大ページ数より大きい値を設定しても良い

# ページのステータスとタイトルをチェック
check_pages(base_url, start_page, end_page)
```
