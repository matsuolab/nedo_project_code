```python
# URLのリストを定義
urls = [
    'https://wordrabbit.jp/grammar/',
    'https://wordrabbit.jp/grammar/unit',
    'https://wordrabbit.jp/grammar/text_structure',
    'https://wordrabbit.jp/grammar/subject',
    'https://wordrabbit.jp/grammar/predicate',
    'https://wordrabbit.jp/grammar/modifier',
    'https://wordrabbit.jp/grammar/connective',
    'https://wordrabbit.jp/grammar/independent',
    'https://wordrabbit.jp/grammar/noun',
    'https://wordrabbit.jp/grammar/verb',
    'https://wordrabbit.jp/grammar/adjective',
    'https://wordrabbit.jp/grammar/adjectival_noun',
    'https://wordrabbit.jp/grammar/postpositional_particle',
    'https://wordrabbit.jp/grammar/auxiliary_verb',
    'https://wordrabbit.jp/grammar/adverb',
    'https://wordrabbit.jp/grammar/pre-noun_adjectival',
    'https://wordrabbit.jp/grammar/conjunction',
    'https://wordrabbit.jp/grammar/interjection',
    'https://wordrabbit.jp/grammar/honorific_all',
    'https://wordrabbit.jp/grammar/honorific',
    'https://wordrabbit.jp/grammar/humble',
    'https://wordrabbit.jp/grammar/polite',
    'https://wordrabbit.jp/grammar/pasive_voice',
    'https://wordrabbit.jp/grammar/causative',
    'https://wordrabbit.jp/grammar/transfer'
]
```

```python
import requests
import csv, json, os, time
from bs4 import BeautifulSoup
from trafilatura import fetch_url, extract


output_dir = ROOT_PATH + "jsonl/"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)


def fetch_all_pages_to_jsonl(jsonl_filename):
    filename = os.path.join(output_dir, jsonl_filename)
    # 抽出したページ数のカウント
    count = 0
    # 全ページ数のカウント
    total_count = 0

    with open(filename, 'w', encoding='utf_8_sig') as file:

      for url in urls:
        total_count += 1
        page_url = url
        downloaded = fetch_url(page_url)
        soup = BeautifulSoup(downloaded, 'html.parser')
        title = soup.title.text if soup.title else ''
        extracted_text = extract(downloaded, include_tables=False, target_lang='ja')

        # "。"を含まないページはフィルタリング（単語のみや英語のページが多いため）
        if not extracted_text or "。" not in extracted_text:
            # print(extracted_text)
            continue

        # jsonl形式で書き込み
        page_data = {'title': title, 'url': page_url, 'text': extracted_text}
        json.dump(page_data, file, ensure_ascii=False)
        file.write('\n')

        count += 1
        print(f"Processed {count} pages.") if count % 1000 == 0 else None
        time.sleep(1)

    # 抽出したページ数を表示
    print(f"Total pages retrieved: {count}/ {total_count}")

fetch_all_pages_to_jsonl('wordrabbit_jp.jsonl')
```

```python
import json
wikibooks = []
with open('/content/drive/MyDrive/Scraping/jsonl/wordrabbit_jp.jsonl', 'r', encoding='utf_8_sig') as f:
    for line in f:
        wikibooks.append(json.loads(line))
```
