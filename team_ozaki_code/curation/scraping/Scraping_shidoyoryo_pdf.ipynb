```python
import os
import json
import fitz  # PyMuPDF

# ルートディレクトリの設定
ROOT_PATH = '/content/drive/MyDrive/Scraping/'

# 処理対象のカテゴリとディレクトリ
categories = ['elem', 'high', 'junior', 'kinder']

# 各カテゴリーのディレクトリをループしてPDFファイルを処理
for category in categories:
    input_dir = os.path.join(ROOT_PATH, 'input', category)
    output_dir = os.path.join(ROOT_PATH, 'jsonl', category)
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # inputディレクトリ内のすべてのPDFファイルを見つける
    for filename in os.listdir(input_dir):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(input_dir, filename)
            jsonl_filename = filename.replace('.pdf', '.jsonl')
            jsonl_path = os.path.join(output_dir, jsonl_filename)

            # PDFからテキストを抽出してJSONLファイルに保存
            doc = fitz.open(pdf_path)
            with open(jsonl_path, 'w', encoding='utf_8_sig') as file:
                for i, page in enumerate(doc, start=1):
                    text = page.get_text("text")
                    page_data = {
                        'title': f"Page {i}",
                        'url': filename,
                        'text': text
                    }
                    json.dump(page_data, file, ensure_ascii=False)
                    file.write('\n')
            print(f"Processed {filename} and saved to {jsonl_path}")
```

```python
import os

# ルートディレクトリの設定
ROOT_PATH = '/content/drive/MyDrive/Scraping/'

# 処理対象のカテゴリとディレクトリ
categories = ['elem', 'high', 'junior', 'kinder']

# 各カテゴリーの行数をカウント
for category in categories:
    output_dir = os.path.join(ROOT_PATH, 'jsonl', category)
    total_lines = 0

    # outputディレクトリ内のすべてのJSONLファイルを見つける
    for filename in os.listdir(output_dir):
        if filename.endswith('.jsonl'):
            jsonl_path = os.path.join(output_dir, filename)

            # ファイルを開いて行数をカウント
            with open(jsonl_path, 'r', encoding='utf-8') as file:
                line_count = sum(1 for line in file)
                total_lines += line_count

    print(f"Total lines in {category}: {total_lines}")
```

```python
import os
import json
import fitz  # PyMuPDF

output_dir = ROOT_PATH + "jsonl/"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def extract_text_by_page_to_jsonl(pdf_path, jsonl_filename):
    doc = fitz.open(pdf_path)
    filename = os.path.join(output_dir, jsonl_filename)
    pdf_file_name = os.path.basename(pdf_path)  # PDFファイル名を取得

    with open(filename, 'w', encoding='utf_8_sig') as file:
        for i, page in enumerate(doc, start=1):
            text = page.get_text("text")
            # JSONL形式で書き込み
            page_data = {
                'title': f"Page {i}",
                'url': pdf_file_name,
                'text': text
            }
            json.dump(page_data, file, ensure_ascii=False)
            file.write('\n')

# PDFファイルのパス
pdf_path = '/content/drive/MyDrive/Scraping/input/20211102-mxt_kyoiku02-100002607_04.pdf'
# JSONLファイル名
jsonl_filename = '20211102-mxt_kyoiku02-100002607_04.jsonl'

# 関数を呼び出してPDFのテキストをJSONLファイルに保存
extract_text_by_page_to_jsonl(pdf_path, jsonl_filename)
```

```python
import fitz  # PyMuPDF

def extract_text_by_page(pdf_path):
    doc = fitz.open(pdf_path)
    page_texts = []  # 各ページのテキストを保存するリスト

    # 各ページをループしてテキストを抽出
    for page in doc:
        text = page.get_text("text")  # ページからテキストを取得
        page_texts.append(text)  # リストにテキストを追加

    return page_texts

# PDFファイルのパス
pdf_path = '/content/drive/MyDrive/Scraping/input/20211102-mxt_kyoiku02-100002607_04.pdf'
page_texts = extract_text_by_page(pdf_path)

# ページごとのテキストを表示（確認用）
for i, text in enumerate(page_texts, 1):
    print(f"Page {i}:\n{text[:500]}")  # 最初の500文字だけ表示する
    print("--------------------------------------------------------------------")
```
