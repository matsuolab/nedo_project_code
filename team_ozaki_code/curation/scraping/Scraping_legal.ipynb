```python
import os
import json
import time
from bs4 import BeautifulSoup
from trafilatura import extract

# ベースディレクトリとHTMLファイル名のリスト
base_dir = '/content/drive/MyDrive/Scraping/input/'
html_files = [
    'kaisha.html', 'douro.html', 'keiho.html', 'kojin.html',
    'minpo.html', 'roudou.html', 'shukujitsu.html', 'kenpo.html'
]

output_dir = base_dir.replace('input', 'jsonl/')  # 出力ディレクトリ
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def fetch_all_pages_to_jsonl(jsonl_filename):
    filename = os.path.join(output_dir, jsonl_filename)
    count = 0
    total_count = 0

    with open(filename, 'w', encoding='utf_8_sig') as file:
        for html_file in html_files:
            file_path = os.path.join(base_dir, html_file)

            # ファイルを開いて読み込む
            with open(file_path, 'r', encoding='utf-8') as html_file:
                html_content = html_file.read()

            total_count += 1
            soup = BeautifulSoup(html_content, 'html.parser')
            title = soup.title.text if soup.title else ''
            extracted_text = extract(html_content, include_tables=False, target_lang='ja')

            # jsonl形式で書き込み
            page_data = {'title': title, 'url': '', 'text': extracted_text}
            json.dump(page_data, file, ensure_ascii=False)
            file.write('\n')

            count += 1
            if count % 100 == 0:  # 100ページごとに進行状況を表示
                print(f"Processed {count} pages.")
            time.sleep(1)  # サーバーに負担をかけないためにスリープ

    # 処理終了後、結果を表示
    print(f"Total pages retrieved: {count} / {total_count}")

fetch_all_pages_to_jsonl('legal.jsonl')
```
